```python
import pandas as pd
```

```python
# 读取文件
data = pd.read_csv("./data/stock_day.csv")
data
```

```python
# 删除一些列，让数据更简单些，再去做后面的操作
data = data.drop(["ma5","ma10","ma20","v_ma5","v_ma10","v_ma20"], axis=1)
data
```

# 索引操作

- 直接使用行列索引(先列后行)

```python
data['open']['2018-02-27']
```

- 结合loc或者iloc使用索引

```python
# 使用loc:只能指定行列索引的名字(先行后列)
data.loc['2018-02-27':'2018-02-22', ['open','low']]
```

```python
# 使用iloc可以通过索引的下标去获取
# 获取前3天数据,前5列的结果
# data.iloc[:3, :5]
data.iloc[:3, [0,3]]
```

- 使用ix组合索引

>Warning:Starting in 0.20.0, the .ix indexer is deprecated, in favor of the more strict .iloc and .loc indexers.

获取行第1天到第4天，['open', 'close', 'high', 'low']这个四个指标的结果

```python
data.index[0:4]
```

```python
data.index[0:4]
```

```python
# 使用ix进行下表和名称组合做引 已经被废弃了
# 第一种:使用loc
data.loc[data.index[0:4], ['open', 'close', 'high', 'low']]
```

```python
data.columns.get_indexer(['open', 'close', 'high', 'low'])
```

```python
# 第二种:使用iloc
data.iloc[0:4, data.columns.get_indexer(['open', 'close', 'high', 'low'])]
```

# DataFrame排序

- DataFrame排序

    使用df.sort_values(by=, ascending=)

    单个键或者多个键进行排序,

    参数：
        by：按照列指定排序参考的键

        ascending:默认升序

        ascending=False:降序

        ascending=True:升序
        
    data.sort_index 对索引进行排序

```python
# 按照单个键进行排序
data.sort_values(by='open',ascending=True)
```

```python
# 按照多个键进行排序
data.sort_values(by=['open','high'],ascending=True)
```

```python
data
```

```python
# 按照索引进行排序
data.sort_index()
```
