# 直方图

## 单列DataFrame直方图

```python
data = np.random.normal(5.0, 3.0, 10000)  # 均值是5，标准差是3
pData = pd.DataFrame(data)  # data的类型转换
pData
```

```python
pData.hist(bins=30)
```

## 多列DataFrame直方图

```python
data = np.random.normal(5.0, 3.0, (10000,100))  # 均值是5，标准差是3
pData = pd.DataFrame(data)  # data的类型转换
pData
```

```python
pData.hist(column=5,bins=30)
```
