# NaN缺失值处理

缺失值如果为NaN Pandas可以直接处理  处理思路:
    
    应用isnull判断是否有缺失数据NaN
    
    应用fillna实现缺失值的填充
    
    应用dropna实现缺失值的删除
    
    应用replace实现数据的替换

```python
# 读取csv文件
data = pd.read_csv('./data/IMDB-Movie-Data.csv')
data
```

- pd断是否有缺失值

  notnull
  
  isnull

```python
# 是否有缺失值NaN
pd.isnull(data)
```

```python
# 按列判断是否有NaN值
pd.isnull(data['Revenue (Millions)'])
```

```python
# 判断是否全部为True
np.all(pd.isnull(data['Revenue (Millions)']))
```

- 填充缺失值

  fillna  填充缺失值
  
  dropna  删除缺失值

```python
# 删除缺失值
data.dropna(axis=1)
```

```python
# 填充缺失值
# inplace:True:会修改原数据，False:不替换修改原数据，生成新的对象
data['Revenue (Millions)'].fillna(data['Revenue (Millions)'].mean(), inplace=True)
```

```python
data
```

```python
np.all(data.notnull())
```

```python
# 每一列NaN都替换
for i in data.columns:
    if not np.all(pd.notnull(data[i])):
        print(i)
        data[i].fillna(data[i].mean(), inplace=True)
```

```python
np.all(data.notnull())
```

# ?缺失值处理

```python
wis = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data")
wis
```

```python
# 把一些其它值标记的缺失值，替换成np.nan inplace为True表示修改原来数据  False不修改原来数据
wis = wis.replace(to_replace='?', value=np.nan,inplace=False)
wis
```

```python
np.all(wis.notnull())
```

```python
# 删除
wis = wis.dropna()
wis
```
